```python
import numpy as np
import matplotlib.pyplot as plt 

import pandas as pd  
import seaborn as sns 

%matplotlib inline
from sklearn.datasets import load_boston
```

```python
boston_dataset = load_boston()
boston = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston['Price'] = boston_dataset.target
boston.head()
```

```python
boston.describe()
```

```python
boston.corr()
```

```python
plt.rcParams["figure.figsize"] = (20,10)
correlation_matrix = boston.corr().round(2)
# annot = True to print the values inside the square
sns.heatmap(data=correlation_matrix, annot=True)
```

The most stronger correlation between price and other features is: price-RM/ price-ZN/ price-CHAS/ price-DIS/ price-B

```python
boston['B'].mean()
```

```python
boston['Price'].mean()
```

```python
fig, ax = plt.subplots()
_ = sns.distplot(boston[boston["RM"] <= 1e5]["RM"], kde=False, 
                  hist_kws={'alpha': 1}, ax=ax)
_ = ax.set(xlabel="RM", ylabel="Price")
```

When the average number of rooms is between 6-7 rooms the price increases (depends on other factors)

```python
fig, ax = plt.subplots()
_ = sns.distplot(boston[boston["ZN"] <= 1e5]["ZN"], kde=False, 
                  hist_kws={'alpha': 1}, ax=ax)
_ = ax.set(xlabel="ZN", ylabel="Price")
```

There is an inverse relationship between ZN and price (whenever lower the ZN the price higher).

```python
fig, ax = plt.subplots()
_ = sns.distplot(boston[boston["B"] <= 1e5]["B"], kde=False, 
                  hist_kws={'alpha': 1}, ax=ax)
_ = ax.set(xlabel="B", ylabel="Price")
```

There is a direct relationship between B and the price (more black people more the price).

```python
fig, ax = plt.subplots()
_ = sns.distplot(boston[boston["CHAS"] <= 1e5]["CHAS"], kde=False, 
                  hist_kws={'alpha': 1}, ax=ax)
_ = ax.set(xlabel="CHAS", ylabel="Price")
```

```python
fig, ax = plt.subplots()
_ = sns.distplot(boston[boston["DIS"] <= 1e5]["DIS"], kde=False, 
                  hist_kws={'alpha': 1}, ax=ax)
_ = ax.set(xlabel="DIS", ylabel="Price")
```

Inverse relationship between distance from employment centers and price (the less space increases the price)
