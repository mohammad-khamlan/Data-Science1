# Trending and publishing date 

```python
dfDate['trending_date'] = pd.to_datetime(dfDate['trending_date'],errors='coerce', format='%y.%d.%m')
dfDate['publish_time'] = pd.to_datetime(dfDate['publish_time'], errors='coerce', format='%Y-%m-%dT%H:%M:%S.%fZ')

dfDate = dfDate[dfDate['trending_date'].notnull()]
dfDate = dfDate[dfDate['publish_time'].notnull()]

dfDate = dfDate.dropna(how='any',inplace=False, axis = 0)

dfDate.insert(4, 'publish_date', dfDate['publish_time'].dt.date)
dfDate['publish_time'] = dfDate['publish_time'].dt.time

my_df_full = df.reset_index().sort_values('trending_date').set_index('video_id')
dfDate = dfDate.reset_index().sort_values('trending_date').drop_duplicates('video_id',keep='last').set_index('video_id')
dfDate[['publish_date','publish_time']].head()
```

```python
dfDate['trending_year'] = dfDate['trending_date'].dt.year 
dfDate['trending _month'] = dfDate['trending_date'].dt.month 
dfDate['trending_day'] = dfDate['trending_date'].dt.day 
dfDate[['publish_year','pulish_month','publish_day']] = dfDate['publish_date'].apply(lambda x: pd.Series(x.strftime("%Y,%m,%d").split(",")))
dfDate
```
